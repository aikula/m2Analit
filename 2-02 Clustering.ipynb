###  1:   K-means

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn import preprocessing

#  
df = pd.read_excel('railway_freight_data.xlsx')

#    
#      ,        
target=['Weight_Tons', 'Cost_USD','Delay_Hours']
X = df[target]
scaler = preprocessing.StandardScaler().fit(X)
X=scaler.fit_transform(X)

#  K-means  3 
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(X)

#  ,     
xLine='Weight_Tons' #  
yLine='Delay_Hours' #  Y
#  
plt.scatter(df[xLine], df[yLine], c=df['Cluster'], cmap='viridis')
plt.title('  K-means')
plt.xlabel('  ()')
plt.ylabel('')
plt.show()
```

### :
- **`KMeans(n_clusters=3)`**   3 .
- **`fit_predict(X)`**          .
- **`plt.scatter()`**        (  ).

#### 

       

```python
df_result=df.groupby('Cluster')[target].mean()
df_result['Count']=df.groupby('Cluster')['Cluster'].count()
df_result
```

###  2:   DBSCAN

```python
from sklearn.cluster import DBSCAN

#  DBSCAN
dbscan = DBSCAN(eps=0.8, min_samples=5)
df['Cluster_DBSCAN'] = dbscan.fit_predict(X)

#  
plt.scatter(df['Weight_Tons'], df['Cost_USD'], c=df['Cluster_DBSCAN'], cmap='plasma')
plt.title('  DBSCAN')
plt.xlabel('  ()')
plt.ylabel('  (USD)')
plt.show()
```

### :
- **`DBSCAN(eps=1000, min_samples=5)`**     DBSCAN,  `eps`     ,  `min_samples`      .

#  eps

    k-               eps   DBSCAN (Density-Based Spatial Clustering of Applications with Noise).

```python
#  k-
from sklearn.neighbors import NearestNeighbors
k = 5  #   
neigh = NearestNeighbors(n_neighbors=k)
nbrs = neigh.fit(X)
distances, indices = nbrs.kneighbors(X)

#  k-
distances = np.sort(distances[:, k-1], axis=0)

#   k-
plt.figure(figsize=(10, 4))
plt.plot(distances)
plt.xlabel(' ,   ')
plt.ylabel(f'{k}-')
plt.title(f' {k}-   eps')
plt.show()
```

###  3:   ()

```python
from scipy.cluster.hierarchy import dendrogram, linkage

#  
Z = linkage(X, method='ward')

plt.figure(figsize=(10, 7))
dendrogram(Z)
plt.title(' ( )')
plt.xlabel('')
plt.ylabel('')
plt.show()
```

### :
- **`linkage()`**         .

```python
from scipy.cluster.hierarchy import fcluster

#      
max_distance = 20  #   ,      
clusters = fcluster(Z, max_distance, criterion='distance')

#     
print(clusters)
```

```python
wcss = []  #       (inertia)

#     k ( 1  10)
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

#    
plt.plot(range(1, 11), wcss, marker='o')
plt.title('      ')
plt.xlabel(' ')
plt.ylabel('WCSS (  )')
plt.show()
```

### :
- **`wcss.append(kmeans.inertia_)`**         k.

2. **   (Average Silhouette Score)**:
   -   ,     .
   -        .

#####        :

```python
from sklearn.metrics import silhouette_score
silhouette_scores = []

#     k ( 2  10)
for i in range(2, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    labels = kmeans.fit_predict(X)
    sil_score = silhouette_score(X, labels)
    silhouette_scores.append(sil_score)

#    
plt.plot(range(2, 11), silhouette_scores, marker='o')
plt.title('      ')
plt.xlabel(' ')
plt.ylabel(' ')
plt.show()
```
